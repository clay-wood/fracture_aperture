```python
def getMeanErr(x):
    avg = np.nanmean(x,axis=0);
    ps_max = np.nanmax(x,axis=0);
    ps_min = np.nanmin(x,axis=0);
    return avg, ps_max, ps_min

p5346C_ps_avg, p5346C_ps_max, p5346C_ps_min = getMeanErr(p5346C_ps)
p5346C2_ps_avg, p5346C2_ps_max, p5346C2_ps_min = getMeanErr(p5346C2_ps)


# p5346C_ps_err = np.nanstd(p5346C_ps,axis=0);
```
