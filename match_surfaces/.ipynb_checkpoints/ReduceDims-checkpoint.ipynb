```python
import numpy as np
# import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import scipy.io
from scipy import signal
# from mpl_toolkits.mplot3d import Axes3D
# import ipympl
```

```python
aperture = np.load('../p5346_aperture.npz')['arr_0']

resolution = 11.758e-6
```

```python
ReduceFactor=5;
# k=0;
# p=0;
# RowLength= 11.758e-6 * ReduceFactor;
RowLength= resolution * ReduceFactor;
```

```python
np.floor(2618/ReduceFactor).astype(int)
```

```python
extent = np.floor(2618/ReduceFactor).astype(int)

ReduceZ = np.zeros((extent,extent))
ReduceZ_oneRow = np.zeros((extent**2))
ReduceX_oneRow = np.zeros((extent**2))
ReduceY_oneRow = np.zeros((extent**2))
```

```python
def Reduction(extent,ReduceZ,ReduceZ_oneRow,ReduceX_oneRow,ReduceY_oneRow):
    k=0;
    p=0;
    
    for aa in range(0,extent-1):
        k = k+1
        l = 0
    
        for bb in range(0,extent-1):
            l = l+1
            p = p+1
            locationk = (aa-1)*ReduceFactor+1;
            locationl = (bb-1)*ReduceFactor+1;
            ReduceZ[k,l] = np.mean(np.mean(aperture[locationk:locationk+1,locationl:locationl+1]));
            ReduceZ_oneRow[p] = np.mean(np.mean(aperture[locationk:locationk+1,locationl:locationl+1]));
            ReduceX_oneRow[p] = l*RowLength
            ReduceY_oneRow[p] = k*RowLength
    
    return ReduceZ,ReduceZ_oneRow,ReduceX_oneRow,ReduceY_oneRow
```

```python
ReduceZ,ReduceZ_oneRow,ReduceX_oneRow,ReduceY_oneRow = Reduction(extent,ReduceZ,ReduceZ_oneRow,ReduceX_oneRow,ReduceY_oneRow)
```

```python
fig = plt.figure(figsize=(6,6),dpi=300)
ax = fig.add_subplot(111)
pos = ax.contourf(ReduceZ, cmap='seismic')
fig.colorbar(pos,label='Aperture Height $(\mu m)$')

plt.ylabel('Length (mm)',size=20)
plt.xlabel('Width (mm)',size=20)
plt.show()
```
