```python
def third_contact_par(aa,bb, aperture, X, Y, row_1, row_2, col_1, col_2, resolution, delta_min, flag):
    # Find the third location

    # flag = np.zeros((len(aperture),len(aperture.T)))
    # for i in range(1,len(aperture)-1):
    #     for j in range(1,len(aperture.T)-1):

    if np.logical_and(np.logical_or(aa != row_1, bb !=col_1), np.logical_or(aa != row_2, bb !=col_2)):
        P1 = np.array([col_1*resolution, row_1*resolution, aperture[row_1,col_1]]).T
        P2 = np.array([col_2*resolution, row_2*resolution, aperture[row_2,col_2]]).T
        P3 = np.array([resolution*bb, resolution*aa, aperture[aa,bb]])
        normal = np.cross(P1-P2, P1-P3)
        intersect = np.dot(normal,P1.T)

        delta = (intersect - np.dot(normal[0:2], np.array([np.ravel(X),np.ravel(Y)]))) / normal[2] - np.ravel(aperture).T

        if np.all(np.ravel(delta) <= delta_min):
            flag[aa,bb] = 1
        else: 
            delta = 0
                    # break

    return aa, bb, delta
```
